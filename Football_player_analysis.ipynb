```python
premier_league_counts = df[df['competition'] == 'Premier League'].groupby(['Player Name', 'competition']).size().reset_index(name='count')

premier_league_counts
```

Here we have sorted out the players who played only for premier league. These are players Aleksandar played 21 times, Harry kane played 30 times and Ivan Toney played 28 times for the English premier league

```python
#goals_assists= df.groupby(["Player Name","Goals","Assists"]).size().reset_index(name='count')

goals_assists= df.groupby(["Player Name"]).agg({'Goals': 'sum', 'Assists': 'sum'}).reset_index()
# goals_assists
```

```python
goals_assists["G_A"] = goals_assists['Goals'] + goals_assists['Assists']
```

```python
goals_assists.columns=['Player Name','Goals','Assists','G_A']
#goals_assists
```

```python
goals_assists= goals_assists.sort_values(by='Goals', ascending=False)
goals_assists
```

Above we got the count of assists and goals scored by each player, 
- Rasmus Hojlund is having 7 goals and 2 assists which is very less.
- Victor osimhen is having highest goals which is 25 and 4 assists.
- Harry kane is scored 24 goals and 4 assists.
- Randal is having high number of assists which is 10 and the goals he has scored is 14.

from the above points we can see that in few Goals are more and in few Assists are more, hypotheticallly there is no perfect insight for us. hence we have created one more variable 'G_A' which is the Addition of Goals and Assists by a player.

- G_A that is the sum of Gaols and Assists.
Now with the help of the 4th column we can say like, Victor Osimhen is having the more 'G_A'points amoong all the players which is 29, then Harry Kane is having 28 points in G_A. Randal is having high number of assists which is 10 and his G_A points are 24.

```python
players_stats= df.groupby('Player Name')['Rating'].agg(['mean', 'median', lambda x: x.mode().iloc[0]]).reset_index()
players_stats.columns = ['Player Name', 'Mean Rating', 'Median Rating', 'Mode Rating']
players_stats.round()
```

```python
player_stats1 = df.groupby('Player Name')['Rating'].agg(['var', 'std']).reset_index()
```

```python
player_stats1.columns = ['Player Name', 'Variance', 'Std Deviation']
player_stats1.round(2)
```

If the std deviation is high, this means that values are typically a long way from mean. if the std deviation is low, values are tend to be close to the mean.

Here the players who are having low std deviation are very consistance in their performance.
